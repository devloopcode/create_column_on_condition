```python
import pandas as pd
import numpy as np

df_laptops = pd.read_csv('laptop_price.csv', encoding='unicode_escape')
```

```python
df_laptops.head()
```

# 1. On 1 condition

```python
np.where(df_laptops['Price_euros'] > 2_000, 'Expensive', 'cheap')
```

```python
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2_000, 'Expensive', 'cheap')
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts(df_laptops['Price_tier'])
```

```python
# create an array based on condition: (screen size > 15)
np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
# set it to a new column
df_laptops['Size'] = np.where(df_laptops['Inches'] > 15, 'Big', 'Small')
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts(df_laptops['Size'])
```

# 2. On 2 or More conditions

```python
df_laptops[(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2_000)].head()
```

```python
df_laptops[((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')) & (df_laptops['Price_euros'] > 2_000)].value_counts('Company')
```

```python
df_laptops[((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')) & (df_laptops['Price_euros'] > 2_000)].head()
```
