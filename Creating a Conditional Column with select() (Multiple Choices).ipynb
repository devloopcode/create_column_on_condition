```python
import pandas as pd
import numpy as np

df_laptops = pd.read_csv('laptop_price.csv', encoding='unicode_escape')
```

```python
df_laptops.head()
```

# 1. Creating a Conditional Column From More Than 2 Choices with select()

```python
conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    df_laptops['Price_euros'] <= 800
]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
df_laptops['Price_tiers'] = np.select(conditions, values, default="Unknown")
```

```python
df_laptops.head(10)
```

```python
df_laptops.value_counts(df_laptops['Price_tiers'])
```

```python
screen_conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <= 14),
    df_laptops['Inches'] <= 12
]

screen_values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
df_laptops['screen_size'] = np.select(screen_conditions, screen_values, default="Unknown")
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts(df_laptops['screen_size'])
```
